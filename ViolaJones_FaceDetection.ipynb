## Viola Jones Face Detection

```python
!pip install opencv-python-headless
import cv2
import matplotlib.pyplot as plt

#Haar cascade for face detection
face_cascade =cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')

image_path= "/content/CV1.jpg"
image= cv2.imread(cv2.samples.findFile(image_path))

#(Viola -Jones works on grayscale images)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Detect faces
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

#Draw rectangles around detected faces \
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 3)

# Output image
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Haar cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

video_path = '/content/v1.mp4'

# Open the video file
cap = cv2.VideoCapture(video_path)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    if not ret:
        break  # If no frame is read, exit the loop (end of video)

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Draw rectangles around detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 3)

    # Display the resulting frame using cv2_imshow in Colab
    cv2_imshow(frame)

    # Wait for a short period, exit if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close any open windows
cap.release()
cv2.destroyAllWindows()
```
